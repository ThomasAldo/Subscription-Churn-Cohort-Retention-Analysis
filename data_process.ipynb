```python
import pandas as pd

print("1. Loading datasets...")
# Load target labels
df_train = pd.read_csv('train_v2.csv')
```

```python
# Load and combine transactions
df_trans_hist = pd.read_csv('transactions.csv')
df_trans_v2 = pd.read_csv('transactions_v2.csv')
df_all_transactions = pd.concat([df_trans_hist, df_trans_v2], ignore_index=True)
```

```python
print("2. Sampling 60k users and merging data...")
# Take exactly 60,000 unique users from the train set to match your resume
sampled_users = df_train.sample(n=60000, random_state=42)

# Merge to get only the transactions for your 60,000 sampled users
portfolio_dataset = pd.merge(sampled_users, df_all_transactions, on='msno', how='inner')
```

```python
print("3. Cleaning columns and formatting dates...")
# Rename columns to be standard business terms
portfolio_dataset.rename(columns={
    'msno': 'customer_id',
    'is_churn': 'churn_status',
    'transaction_date': 'payment_date',
    'membership_expire_date': 'expire_date',
    'actual_amount_paid': 'amount'
}, inplace=True)
```

```python
# Convert integer dates (e.g., 20170228) to actual datetime formats for SQL/Power BI
portfolio_dataset['payment_date'] = pd.to_datetime(portfolio_dataset['payment_date'], format='%Y%m%d')
portfolio_dataset['expire_date'] = pd.to_datetime(portfolio_dataset['expire_date'], format='%Y%m%d')

# Keep only the columns we actually need for the project
columns_to_keep = ['customer_id', 'churn_status', 'payment_date', 'expire_date', 'amount', 'is_auto_renew', 'is_cancel']
portfolio_dataset = portfolio_dataset[columns_to_keep]
```

```python
print("4. Exporting master dataset...")
portfolio_dataset.to_csv('portfolio_subscription_master.csv', index=False)

unique_users = portfolio_dataset['customer_id'].nunique()
total_rows = len(portfolio_dataset)
print(f"Success! Master dataset generated.")
print(f"Total Unique Customers: {unique_users}")
print(f"Total Transaction Rows: {total_rows}")
```
